# Method 1:  Determine k for spring 1 and 2

```python
import physplot as phys
import numpy as np

#smallest k spring
m1 = np.array([50,100,150,200,250,300])/1000
x1 = np.array([17,33.5,50,66,83,98.5])/100
errm1 = np.array([.2,.2,.2,.2,.2,.2]) #in meters

#largest k spring
m2 = np.array([50,100,150,200,250,300])/1000
x2 = np.array([5,9.5,13,17,21.5,25.5])/100
errm2 = np.array([.6,.6,.6,.6,.5,.5]) #in meters

#call the plotfit function from the physplot.py code
#You can open that code to see how it works
#This is what the plotfit function looks like.  Pass
#in the parameters in the correct order
#plotfit(x,y,dy,title,xlabel,ylabel,slopeunit)

phys.plotfit(x1,m1*9.8,errm1*9.8,'Small k spring','x(m)','F(N)','N/m')
phys.plotfit(x2,m2*9.8,errm2*9.8,'Largest k spring','x(m)','F(N)','N/m')
```
