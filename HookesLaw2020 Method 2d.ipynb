# Method 2:  Plotting *T<sup>2</sup>* vs *m* for k

```python
import physplot as phys
import numpy as np

#smallest k spring
m3 = np.array([50,100,150,200,250,300])/1000
t3 = np.array([.1,9.5,13,17,21.5,25.5])/10

#largest k spring
m4 = np.array([50,100,150,200,250,300])/1000
t4 = np.array([.1,9.5,13,17,21.5,25.5])/10

errt = .3/10 #based on reaction time

errt3_sq = 2*t3*errt
errt4_sq = 2*t4*errt

#call the plotfit function from the physplot.py code
#You can open that code to see how it works
#This is what the plotfit function looks like.  Pass
#in the parameters in the correct order
#plotfit(x,y,dy,title,xlabel,ylabel,slopeunit)

phys.plotfit(m3,t3**2,errt3_sq,'Small k spring','mass(kg)','$T^2(s^2)$','s^2/kg')
phys.plotfit(m4,t4**2,errt4_sq,'Large k spring','mass(kg)','$T^2(s^2)$','s^2/kg')
```
